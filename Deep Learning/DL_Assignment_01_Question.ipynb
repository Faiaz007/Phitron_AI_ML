# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.




## Write Answer 01:
Answer: A perceptron is the simplest form of an artificial neuron and the basic building block of neural networks. It is a linear binary classifier which is designed to take multiple input features, process them, and produce a single binary output typically 0 or 1.
It mimics how a biological neuron works. It receives signals inputs, weigths their importance, sums them up, and decides whether to activate or not based on whether the total exceeds a certain threshold.

Three main components:


1.   Inputs(x1,x2..): The data fed into the perceptron.
2.   Weights(w1,w2..): Numbers that show how important each input is. They get adjusted during learning.
3.   Bias (b): A constant that shifts the decision boundary so the perceptron can better fit the data.

# Question 02: [ Marks 10 ]

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.




## Write Answer 02:

A decision boundary in a perceptron is the boundary that separates input data into different classes based on the perceptron’s prediction. It represents the set of points where the perceptron is unsure which class to assign.

A perceptron computes a weighted sum of inputs:

w₁x₁ + w₂x₂ + b

The output depends on the sign of this value. If the result is positive, the perceptron predicts one class; if negative, it predicts the other.

The decision boundary is defined by setting this equation equal to zero:

w₁x₁ + w₂x₂ + b = 0

Geometrically, in a 2-dimensional feature space, this equation represents a straight line. This line divides the plane into two regions. Points on one side of the line satisfy w₁x₁ + w₂x₂ + b > 0 and belong to one class, while points on the other side satisfy w₁x₁ + w₂x₂ + b < 0 and belong to the other class.

The weight vector (w₁, w₂) determines the orientation of the line, and the bias b shifts the line’s position.




# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.



## Write Answer 03:
The perceptron is called a linear classifier because it separates data using a linear decision boundary, such as a straight line in two dimensions or a plane in higher dimensions. Its decision rule is based on a weighted sum of inputs plus a bias, and the classification depends only on whether this sum is positive or negative. This always produces a linear boundary.

Linear separability means that two classes of data can be separated completely by a single straight line (or hyperplane). If such a line exists, the data is said to be linearly separable, and a perceptron can learn it successfully.

Logical gates like AND and OR are linearly separable. For the AND gate, the output is 1 only when both inputs are 1. These input–output pairs can be separated from the 0 outputs using a single straight line. Similarly, for the OR gate, the inputs that produce output 1 can also be separated from those that produce output 0 with one straight line. Therefore, a single perceptron can correctly learn AND and OR functions.

However, the XOR gate is not linearly separable. In XOR, the output is 1 when the inputs are different and 0 when they are the same. No single straight line can separate the 1 outputs from the 0 outputs in the input space. Because a perceptron can only create linear decision boundaries, it fails to learn XOR.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
The perceptron weight update rule is used to adjust the model so that it makes fewer mistakes on future predictions. For a training example with input vector x, target label t, predicted output y, and learning rate η, the update rule is:

w(new) = w(old) + η (t − y) x
b(new) = b(old) + η (t − y)

Intuitively, the perceptron learns by correcting its errors. If the perceptron predicts the correct class, then t − y = 0, so no update is needed because the current weights already classify that example correctly. Changing them could actually harm correct behavior.

If the perceptron makes a mistake, the update pushes the decision boundary in the right direction. When the true label is 1 but the prediction is 0, the weights are increased in the direction of the input, making it more likely that similar inputs will be classified as 1 next time. When the true label is 0 but the prediction is 1, the weights are decreased, reducing the influence of that input on future predictions.

Weights are updated only when the prediction is incorrect because learning happens only when there is an error to correct. Updating weights for correct predictions is unnecessary and can destabilize the decision boundary. This error-driven update mechanism ensures that the perceptron converges for linearly separable data.


# Question 05: [ Marks 10 ]

Why does a single-layer perceptron fail to solve the XOR problem?
What does this limitation tell us about the need for multilayer neural networks?

## Write Answer 05:
A single-layer perceptron fails to solve the XOR problem because XOR is not linearly separable. In the XOR function, inputs that produce output 1 and inputs that produce output 0 cannot be separated by a single straight line in the input space. A single-layer perceptron can only form one linear decision boundary, so it cannot correctly classify all XOR input combinations.

This limitation shows that single-layer perceptrons are only capable of solving linearly separable problems. To handle problems like XOR, the model needs multiple layers. Multilayer neural networks introduce hidden layers and nonlinear transformations, which allow the network to combine multiple linear boundaries and learn complex, non-linear decision surfaces. This is why multilayer networks are essential for solving real-world problems that are not linearly separable.

# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:
The perceptron learning rule, also called the weight adjustment method, is the procedure used to train a perceptron by iteratively updating its weights so that it correctly classifies training data. For each training example, the perceptron computes a weighted sum of the inputs and produces an output. If the output matches the target label, no change is made. If the output is incorrect, the weights and bias are adjusted to reduce the error.

The learning rule is expressed as:

w(new) = w(old) + η (t − y) x
b(new) = b(old) + η (t − y)
where η is the learning rate, t is the true label, y is the predicted output, and x is the input vector.

This rule works by moving the decision boundary in a direction that improves classification of the misclassified point. Increasing or decreasing a weight changes the slope or position of the decision boundary.

Adjusting one weight can affect the classification of other points because all input features share the same set of weights. When a weight is changed, the decision boundary shifts globally, not locally. This shift may fix the classification of one misclassified point but can also move other points closer to or farther from the boundary, potentially changing their predicted class. This interaction is why perceptron learning is an iterative process and may require multiple passes over the data to reach a stable solution.

The main limitations of a single-layer perceptron come from its simplicity. First, it can only create a linear decision boundary. This means it can solve only linearly separable problems and completely fails on non-linear problems such as XOR. Many real-world datasets are not linearly separable, so a single-layer perceptron is often too weak for practical use.

Second, a single-layer perceptron has very limited representational power. It cannot learn complex patterns, feature interactions, or hierarchical relationships in data. It treats all inputs in a flat manner and cannot automatically extract higher-level features.

Third, it relies on a hard threshold activation function, which makes it unsuitable for learning smooth decision boundaries or probabilities. This also limits its flexibility and learning behavior.

Using more than one layer, as in a multilayer perceptron (MLP), helps overcome these limitations by introducing hidden layers and non-linear activation functions. Each hidden layer can learn intermediate representations or features from the input. By stacking layers, the network can combine multiple linear transformations with non-linearities, allowing it to form complex, non-linear decision boundaries.
