# Question 02: Manual Gradient Calculation

## Given

Linear model:

$$
z = w x + b
$$

Prediction using sigmoid:

$$
y_{pred} = \sigma(z) = \frac{1}{1 + e^{-z}}
$$

Binary Cross Entropy Loss:

$$
L = - \left[ y \log(y_{pred}) + (1-y)\log(1-y_{pred}) \right]
$$



# a) Compute $\frac{dL}{dy_{pred}}$

Start with the loss function:

$$
L = - \left[ y \log(y_{pred}) + (1-y)\log(1-y_{pred}) \right]
$$

### Step 1: Differentiate the first term

$$
\frac{d}{dy_{pred}} \left( y \log(y_{pred}) \right)
$$

Since $y$ is constant:

$$
= y \cdot \frac{1}{y_{pred}}
$$



### Step 2: Differentiate the second term

$$
\frac{d}{dy_{pred}} \left( (1-y)\log(1-y_{pred}) \right)
$$

Using the chain rule:

$$
= (1-y)\cdot \frac{1}{1-y_{pred}} \cdot (-1)
$$



### Step 3: Combine both terms

$$
\frac{dL}{dy_{pred}} =
-\left[
\frac{y}{y_{pred}} -
\frac{1-y}{1-y_{pred}}
\right]
$$

Final result:

$$
\frac{dL}{dy_{pred}}
=
-\frac{y}{y_{pred}} + \frac{1-y}{1-y_{pred}}
$$



# b) Compute $\frac{dy_{pred}}{dz}$

Sigmoid function:

$$
y_{pred} = \sigma(z)
$$

Derivative of sigmoid:

$$
\frac{d}{dz}\sigma(z) = \sigma(z)(1-\sigma(z))
$$

Since $y_{pred} = \sigma(z)$:

$$
\frac{dy_{pred}}{dz} = y_{pred}(1-y_{pred})
$$



# c) Show that $\frac{dL}{dz} = y_{pred} - y$

Using the chain rule:

$$
\frac{dL}{dz}
=
\frac{dL}{dy_{pred}} \cdot \frac{dy_{pred}}{dz}
$$

Substitute results from (a) and (b):

$$
\frac{dL}{dz}
=
\left(
-\frac{y}{y_{pred}} + \frac{1-y}{1-y_{pred}}
\right)
\cdot
y_{pred}(1-y_{pred})
$$



### Step 1: Multiply the first term

$$
-\frac{y}{y_{pred}} \cdot y_{pred}(1-y_{pred})
$$

Cancel $y_{pred}$:

$$
= -y(1-y_{pred})
$$



### Step 2: Multiply the second term

$$
\frac{1-y}{1-y_{pred}} \cdot y_{pred}(1-y_{pred})
$$

Cancel $(1-y_{pred})$:

$$
= (1-y)y_{pred}
$$



### Step 3: Combine

$$
\frac{dL}{dz}
=
-y(1-y_{pred}) + (1-y)y_{pred}
$$

Expand:

$$
= -y + yy_{pred} + y_{pred} - yy_{pred}
$$

Cancel $yy_{pred}$:

$$
= y_{pred} - y
$$



# Final Result

$$
\frac{dL}{dz} = y_{pred} - y
$$

# b) Compute dL/dx Manually

First write the loss function.

$$
L = y_1 + y_2
$$

Substitute the equations.

$$
L = x^2 + 4x
$$

Now compute the derivative with respect to **x**.

### Step 1: derivative of \(x^2\)

$$
\frac{d}{dx}(x^2) = 2x
$$

### Step 2: derivative of \(4x\)

$$
\frac{d}{dx}(4x) = 4
$$

### Step 3: combine them

$$
\frac{dL}{dx} = 2x + 4
$$



### Evaluate at \(x = 3\)

$$
\frac{dL}{dx} = 2(3) + 4
$$

$$
= 6 + 4
$$

$$
= 10
$$

Final result:

$$
\frac{dL}{dx} = 10
$$

# c) How to Compute Using Autograd

Autograd automatically computes derivatives by tracking operations in a computation graph.

Example using **PyTorch Autograd**.

```python
import torch

#define x
x= torch.tensor (3.0, requires_grad=True)

#Forward Pass computation
y1 = x**2
y2 = 4*x
L = y1 + y2

#Backward Pass
L.backward()

#gradient
print(x.grad)
```

# Question 05: [ Marks 10 ]

Create two tensors:
w = 2.0 (requires_grad=True) <br>
b = 1.0 (requires_grad=True) <br>

Let input x = 3.0. <br>
Compute z = w*x + b. <br>
Use Sigmoid to get y_pred.
 <br>

Define Binary Cross Entropy manually (without built-in loss). <br>
Compute gradients using backward(). <br>
Print gradients of w and b. <br>

Write full code.

## Write Answer 05:

```python
#define wieghts and biases
w = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)


#Define Input X
x= torch.tensor(3.0)

#Forward Pass
z = w*x + b


#Sigmoid
y_pred = torch.sigmoid(z)

#Binary Cross Entropy Label
y = torch.tensor(1.0)

#Binary Cross Entropy Manually

loss = -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred))

#Backward Pass
loss.backward()


print("Gradient of w:", w.grad)
print("Gradient of b:", b.grad)
```

Create tensor w = 2.0 (requires_grad=True). <br>

Print w.grad. <br>

Do not clear gradients. <br>
Explain what you observe. <br>

## Write Answer 07:



```python
# Create tensor with gradient tracking
w = torch.tensor(2.0, requires_grad=True)

# First loss
loss1 = w**2
loss1.backward()

print("Gradient after first backward:", w.grad)

# Second loss
loss2 = 3*w
loss2.backward()

print("Gradient after second backward:", w.grad)
```
