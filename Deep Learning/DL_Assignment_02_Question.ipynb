# Question 02: [ Marks 10 ]

What is forward propagation in an artificial neural network? Explain step by step how data moves from the input layer to the output layer.

## Write Answer 02:
Forward propagation is the process  of passing input data through a neural network to generate an output. The input layer recieves fetaure values and forward them to the hidden layer.During this phase, the network calculates an answer but does not learn or adjust itself.

Steps:

1. Input Layer (Data Enters): Raw data (like image pixels or housing specs) is fed into the network's starting neurons as numbers.
2. Weighted Sum (The Math): As data moves to the first hidden layer, each input is multiplied by a Weight (which determines its importance). These numbers are added together, and a Bias (a baseline threshold) is added.
3. Activation Function (The Decision): This sum is passed through an Activation Function (like ReLU or Sigmoid). This function acts as a gatekeeper, deciding how strongly the neuron should "fire" and pass its signal forward.
4. Hidden Layers (The Relay Race): The output from the first layer becomes the input for the next. Steps 2 and 3 repeat across all hidden layers, allowing the network to build upon simple patterns to recognize complex ones.
5. Output Layer (The Final Prediction): The data reaches the final layer, which performs one last calculation to deliver the ultimate result such as a probability (e.g., "85% chance it's a dog") or a raw value (e.g., a predicted house price).


# Question 03: [ Marks 15 ]
Why do we need activation functions in neural networks? What would happen if we only used linear transformations without any activation function?

## Write Answer 03:
Activation functions are required in neural networks to introduce non-linearity. Most real world problems are non-linear, and linear models cannot capture complex patterns. If no activation function is used, each layer performs only a linear transformation. Stacking multiple linear layers is mathematically equivalent to a single linear layer.Because the network acts as a single linear equation, it would only be capable of solving simple problems that can be separated by a straight line. As a result, increasing network depth would not increase learning capacity. The network would fail to learn complex decision boundaries. Therefore, activation functions are essential for neural networks to model non-linear relationships.

# Question 04: [ Marks 15 ]

Explain the sigmoid activation function. Where is it used? What are its advantages and disadvantages? What problem does it cause during training?

## Write Answer 04:
The sigmoid activation function maps any real input to a value between 0 and 1 and is defined as
𝜎(𝑥)=1 / (1 + e⁻ˣ)
	​
.It is mainly used in the output layer of binary classification networks to produce probabilities. Its advantages are smoothness, differentiability, and bounded output, which suit gradient-based optimization. However, it saturates for very large or small inputs, causing gradients to become very small. This leads to the vanishing gradient problem during training, slowing or stopping learning. Because of this, sigmoid is rarely used in hidden layers of modern deep networks.

# Question 05: [ Marks 10 ]
Explain the softmax activation function in simple words. Why is it used in multi-class classification problems? How is it different from sigmoid?

## Write Answer 05:
The softmax activation function converts a vector of raw scores (logits) into probabilities that sum to 1. It is defined as  Softmax(zᵢ) = e^(zᵢ) / Σ e^(zⱼ). Softmax is mainly used in multi-class classification problems, where the model must assign a probability to each class, and the sum of all probabilities should be 1. Unlike sigmoid, which outputs independent probabilities between 0 and 1 for each class, softmax considers all classes together, producing mutually exclusive probabilities. This makes softmax ideal when only one class can be the correct output.

# Question 06: [ Marks 20 ]

Explain the structure of an Artificial Neural Network in detail. Describe the role of the input layer, hidden layers, and output layer. How do weights, biases, and activation functions work together to transform input data into predictions? Illustrate your explanation with a simple example.

## Write Answer 06:
An Artificial Neural Network (ANN) is a computational model inspired by the human brain, consisting of layers of interconnected nodes called neurons. It has three main types of layers: the input layer, hidden layers, and output layer. The input layer receives raw feature values from the dataset and passes them to the next layer without modification. Hidden layers process the data by applying weights to each input, adding a bias, and then applying an activation function to introduce non-linearity. Activation functions allow the network to learn complex patterns rather than just linear relationships. The output layer produces the final prediction by applying similar computations; in classification problems, this might be a softmax or sigmoid function depending on the task.

Weights determine the importance of each input, while biases allow neurons to shift the activation function, enabling better fitting to the data. Together with activation functions, they transform the input into meaningful outputs. For example, consider a network that predicts whether a student passes an exam based on study hours and sleep hours. The inputs are multiplied by weights, biases are added, and the activation function determines if the neuron “fires.” After passing through hidden layers, the output neuron produces a probability of passing the exam, which can be interpreted as the prediction.

Thus, an ANN learns by adjusting weights and biases during training to minimize error, using forward propagation to compute predictions and backpropagation to update parameters.
