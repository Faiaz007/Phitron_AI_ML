## NDArray

```python
import numpy as np
#1d array
arr=np.array([1,2,3,4,5])


print(type(arr))
print(arr.ndim)

#2d array

arr2 = np.array([ [1,2,3], # row1
                  [4,5,6] #row2
                ])

print(arr2*5)
print(type(arr2))
print(arr2.ndim)



arr3 = np.array([
    #1st floor
    [ [1,2,3], # row1
      [4,5,6]  ], #row2 

    #2nd floor

     [[1,2,3], # row1
      [4,5,6]  ], #row2  


    #3rd floor
      [[1,2,3], # row1
      [4,5,6]  ] #row2     
])
print(arr3*5)
print(type(arr3))
print(arr3.ndim)
```

```python
#Dimension
print(arr.ndim)
print(arr2.ndim)
print(arr3.ndim)
print("\n")
#shape
print(f"Array 1 data Shape:{arr.shape}")
print(f"Array 2 data Shape:{arr2.shape}")
print(f"Array 3 data Shape:{arr3.shape}")
print("\n")
#data type
print(f"Array 1 data Type:",arr.dtype)
print(f"Array 2 data Type:",arr2.dtype)
print(f"Array 3 data Type:",arr3.dtype)

print("\n")
#size
print(f"Array 1 data size:{arr.size}")
print(f"Array 2 data size:{arr2.size}")
print(f"Array 3 data size:{arr3.size}")
```
