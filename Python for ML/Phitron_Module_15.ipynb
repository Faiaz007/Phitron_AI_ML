```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python

hours_1 = [2 , 3 , 4, 1 , 5 , 7 , 3] 
hours_2 = [ 1 , 4 , 3 , 5 ,7 ,2 ,1]
days = [1 ,2 , 3 ,4 ,5 ,6 ,7] 

plt.xlabel('Days')
plt.ylabel('Study Hours') 
plt.title('Trend of a student study time over a week')

plt.plot(days,hours_1)
plt.plot(days,hours_2) 
```

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Years')
plt.ylabel('Count of enrollment')
plt.title('Student enrollment over 8 years')

plt.plot(df['Year'],df['Programming'])
```

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'])
plt.plot(df['Year'] ,df['Digital Marketing'])
```

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming', color ='#11b86f',linewidth=3,linestyle='dashed',marker='o',markersize=7)
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing', color='black',linewidth=3,linestyle='dashed' )

plt.legend() 
```

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming', color ='#11b86f',linewidth=3,linestyle='dashed',marker='o',markersize=7)
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing', color='red',linewidth=3,linestyle='dashed',marker='o',markersize=7 )

plt.legend() 
plt.grid() 

plt.show() 
```
