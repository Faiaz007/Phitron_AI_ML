```python
a = np.random.randint(1,100, size=(2,3))
b = np.random.randint(20,100, size=(2,3))

print(a)
print("\n")
print(b)
```

### Concatenate

```python
con_row = np.concatenate((a,b), axis=0) #row wise
print(con_row)
con_col = np.concatenate((a,b), axis=1) #col wise  
print("\n")
print(con_col)
```

### Transpose

```python
mat = np.array([[10,20,30],[30,40,50]])

tarnspose = mat.T #row to col

print(tarnspose)
```

```python

x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4  ,5 ]) 

add = np.add(x,y) 
print(add)

sub = x-y 
print(sub)

mul = x*y 
print(mul) 


div = x/y 
print(div) 

remainder = x % y

print(remainder) 
```

```python
# mathematical function 

# trigonometry 

# sin , cos , tan takes values in radian 

sin_val = np.sin(x)
print(sin_val) 

cos_val = np.cos(x)
print(cos_val) 

deg_con = np.rad2deg(x)
print(deg_con)

# deg2rad 
```

```python
x = np.array([ 10 , 8 , 16 , 100 ])
base_10_log_val = np.log10(x)
print(base_10_log_val)


base_2_log_value = np.log2(x) 
print(base_2_log_value)

sqrt= np.sqrt(x)
print(sqrt)
```

```python
# different 
s = np.sum(x)
print(s)

#prefix sum 
cumul =np.cumulative_sum(x)
print(cumul)
```

```python
#inplace 
x = np.array([ 10 , 8 , 16 , 100 ])

z = x.copy()

print(z) 

z.sort()

print(z)
```

```python
# copy sorting
print(x)
sort_arr = np.sort(x) 

print(sort_arr)

print(x)
```

```python
# 2d array sorting 
mat = np.array([[10,3,5],[8,4,9]])

# horizon axis = 1 

hor_sort = np.sort(mat,axis=1)
print(hor_sort)
print("\n")
vert_sort = np.sort(mat,axis=0)
print(vert_sort)
```
