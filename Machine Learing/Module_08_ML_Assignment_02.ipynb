### 5. Handling Missing Values (15 marks)
**Tasks:**
1. Show the count of missing values in each column.
2. Decide how to handle missing values for:
   - `Age` (numeric)
   - `Embarked` (categorical)
   - `Cabin` (many missing values)
3. Implement your chosen strategy in code.
4. Show missing value counts again to confirm.
5. Explain your choices in **3–4 sentences**.


```python
# 5. Handling Missing Values

print("Missing values before:") #[Task 1 is done for you]
print(df.isna().sum())

# Drop Cabin (Complete this code block)
if 'Cabin' in df.columns:
    df = df.drop('Cabin', axis=1)
    print("\n'Cabin' column dropped.")
else:
    print("\n'Cabin' already removed earlier.")

# Fill Age (Complete this code block)
df['Age'] = df['Age'].fillna(df['Age'].median())
# Fill Embarked (Complete this code block)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print("\nMissing values after handling:")
print(df.isna().sum())
```

#### Explanation of your strategy
- Why you dropped or kept `Cabin`: Cabin has 77% missing values (687/891) and 147 unique values too sparse and noisy. Dropping it is the standard practice in Titanic analysis.
- Why you used median/mean for `Age`:  Age is slightly right-skewed with some high-age outliers. Median is robust and not influenced by extreme values, unlike the mean its best choice.
- Why you used mode or another method for `Embarked`: Only 2 missing values, and 'S' (Southampton) is by far the most common port (72%). Filling with the mode introduces minimal bias.

### 6. Encoding Categorical Variables (15 marks)
Focus on the following categorical features:
- `Sex`
- `Embarked`
- `Pclass` (treat as categorical)

**Tasks:**
- Use `OneHotEncoder` from sklearn to encode these features.
- Keep `Survived` and the main numeric features (`Age`, `Fare`, `SibSp`, `Parch`).
- Show the shape and head of the encoded DataFrame.


```python
# 6. Encoding Categorical Variables

df_enc = df.copy()

# Select features for encoding and keep important numeric ones
cat_features = ["Sex", "Embarked", "Pclass"]
numeric_features = ["Age", "Fare", "SibSp", "Parch"]
target_col = "Survived"

# One hot encode categorical variables [You need to do]
# Initialize OneHotEncoder (drop='first' to avoid multicollinearity)
ohe = OneHotEncoder(sparse_output=False, drop='first', dtype=int)

# Fit and transform the categorical features
encoded_array = ohe.fit_transform(df_enc[cat_features])

# Get the new column names
encoded_columns = ohe.get_feature_names_out(cat_features)

# Create DataFrame with encoded features
encoded_df = pd.DataFrame(encoded_array, columns=encoded_columns, index=df_enc.index)

# Build final encoded DataFrame
df_model = pd.concat([
    df_enc[[target_col] + numeric_features],
    encoded_df
], axis=1)

print("Shape after encoding:", df_model.shape)
display(df_model.head())
```

---
### 7. Scaling Numeric Features (15 marks)
**Tasks:**
- Use `StandardScaler` on the numeric features: `Age`, `Fare`, `SibSp`, `Parch`.
- Plot boxplots of these features **before** and **after** scaling.
- Write **two sentences** explaining how scaling changes the numeric values and why it is useful before training some ML models.


```python
# 7. Scaling Numeric Feature

score_cols = ["Age", "Fare", "SibSp", "Parch"]

# Boxplot BEFORE scaling
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.boxplot(data=df_model[score_cols])
plt.title("Before Scaling")
plt.ylabel("Original values")

# Apply StandardScaler
scaler = StandardScaler()
scaled_array = scaler.fit_transform(df_model[score_cols])

# Create scaled DataFrame (keep column names)
df_scaled = df_model.copy()
df_scaled[score_cols] = scaled_array

# Boxplot AFTER scaling
plt.subplot(1, 2, 2)
sns.boxplot(data=df_scaled[score_cols])
plt.title("After StandardScaler")
plt.ylabel("Standardized values")

plt.tight_layout()
plt.show()

print("Shape after scaling:", df_scaled.shape)
```

#### Your observations on scaling
- Sentence 1: StandardScaler transformed all numeric features to have a mean of approximately 0 and standard deviation of 1, making Age, Fare, SibSp, and Parch directly comparable despite their originally very different scales and ranges.
- Sentence 2:  This is essential before training models like Logistic Regression, SVM, KNN, or Neural Networks because these algorithms use distance or gradient calculations that are highly sensitive to feature scales without scaling, Fare would dominate the learning process simply due to its larger magnitude.
